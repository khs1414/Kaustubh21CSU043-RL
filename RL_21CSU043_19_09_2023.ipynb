```python
import numpy as np

# Define the number of arms and their true reward probabilities
num_arms = 5
true_probabilities = [0.9, 0.8, 0.7, 0.6, 0.5]

# Initialize variables to keep track of arm statistics
num_pulls = np.zeros(num_arms)  # Number of times each arm has been pulled
sum_rewards = np.zeros(num_arms)  # Total rewards obtained from each arm

# Define the number of rounds (time steps)
num_rounds = 1000

# Main Thompson Sampling loop
for t in range(num_rounds):
    # Sample a random value for each arm's reward probability
    sampled_probabilities = np.random.beta(sum_rewards + 1, num_pulls - sum_rewards + 1)

    # Choose the arm with the highest sampled probability
    chosen_arm = np.argmax(sampled_probabilities)

    # Simulate pulling the chosen arm and receiving a reward
    reward = np.random.binomial(1, true_probabilities[chosen_arm])

    # Update arm statistics
    num_pulls[chosen_arm] += 1
    sum_rewards[chosen_arm] += reward

    # Print the chosen arm and reward at each time step (optional)
    print(f"Round {t+1}: Chose Arm {chosen_arm + 1}, Reward = {reward}")

# Print the final estimated reward probabilities for each arm
estimated_probabilities = sum_rewards / num_pulls
for i in range(num_arms):
    print(f"Estimated Probability for Arm {i+1}: {estimated_probabilities[i]}")

# Print the arm with the highest estimated probability (the best arm)
best_arm = np.argmax(estimated_probabilities)
print(f"The best arm is Arm {best_arm + 1} with an estimated probability of {estimated_probabilities[best_arm]}")
```
