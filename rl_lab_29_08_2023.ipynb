```python
import numpy as np
import matplotlib.pyplot as plt

# Number of arms (bandits)
num_arms = 5

# True mean rewards for each arm (unknown to the algorithm)
true_means = np.random.normal(0, 1, num_arms)

# Initialize variables for UCB algorithm
num_iterations = 1000
num_selections = np.zeros(num_arms)
sum_rewards = np.zeros(num_arms)
ucb_rewards = []

# Initialize variables for random sampling
random_rewards = []

# Simulate UCB algorithm and random sampling
for t in range(1, num_iterations + 1):
    arm_to_pull_ucb = np.argmax(sum_rewards / (num_selections + 1e-6)
    + np.sqrt(2 * np.log(t) / (num_selections + 1e-6)))
    reward_ucb = np.random.normal(true_means[arm_to_pull_ucb], 1)
    sum_rewards[arm_to_pull_ucb] += reward_ucb
    num_selections[arm_to_pull_ucb] += 1
    ucb_rewards.append(reward_ucb)

    arm_to_pull_random = np.random.choice(num_arms)
    reward_random = np.random.normal(true_means[arm_to_pull_random], 1)
    random_rewards.append(reward_random)

# Calculate cumulative rewards for both algorithms
cumulative_ucb_rewards = np.cumsum(ucb_rewards)
cumulative_random_rewards = np.cumsum(random_rewards)

# Plot results
plt.plot(range(1, num_iterations + 1), cumulative_ucb_rewards, label='UCB Algorithm')
plt.plot(range(1, num_iterations + 1), cumulative_random_rewards, label='Random Sampling')
plt.xlabel('Iterations')
plt.ylabel('Cumulative Reward')
plt.legend()
plt.title('Multi-Armed Bandit: UCB vs Random Sampling')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the order of sockets (worst to best)
socket_order = [2, 1, 3, 5, 4]

# Calculate the mean rewards for each socket
mean_rewards = [order * 2 + 2 for order in socket_order]

# Simulate the rewards obtained from each socket
num_iterations = 1000
socket_rewards = [np.random.normal(mean, 1, num_iterations) for mean in mean_rewards]

# Create a violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=socket_rewards)
plt.xlabel('Socket')
plt.ylabel('Reward')
plt.title('Reward Distribution of Each Socket')
plt.xticks(range(len(socket_order)), socket_order)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the order of sockets (worst to best)
socket_order = [2, 1, 3, 5, 4]

# Calculate the mean rewards for each socket
mean_rewards = [order * 2 + 2 for order in socket_order]

# Simulate the rewards obtained from each socket
num_iterations = 1000
socket_rewards = [np.random.normal(mean, 1, num_iterations) for mean in mean_rewards]

# Create a density plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data=socket_rewards, common_norm=False)
plt.xlabel('Socket Output')
plt.ylabel('Density')
plt.title('Density Plot of Socket Outputs')
plt.legend(labels=socket_order)
plt.show()
```
